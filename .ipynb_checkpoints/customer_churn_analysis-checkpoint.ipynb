# Customer Churn Analysis Pipeline

This notebook demonstrates a complete data analysis pipeline for analyzing customer churn. It covers database setup, data loading, SQL-based analysis, Create Features, rule-based risk modeling, and visualization.

### Pipeline Steps:
1. **Database Setup**: Initialize SQLite database from source CSVs.
2. **Data Loading**: Load data from SQLite into Pandas DataFrames.
3. **SQL Analysis**: Calculate churn rate and spending behavior.
4. **Create Features**: Aggregate customer-level metrics.
5. **Risk Modeling**: Rule-based churn risk scoring.
6. **Visualization**: Charts for interpretation.

## 1. Imports and Setup

### What’s happening in this section

This section imports the main Python libraries used throughout the project.

I use:
* **SQLite** to store and query structured data using SQL  
* **Pandas** to work with tables and perform data analysis  
* **NumPy** for basic numerical operations  
* **Matplotlib** to create visualizations  

These tools are commonly used together in data analytics roles, and they form the foundation of the analysis in this notebook.

```python
import os
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

## 2. Database Setup

### What’s happening here

Instead of working directly with CSV files, this project loads the data into a **SQLite database**.  
The goal is to simulate how data is usually stored and accessed in a real analytics environment.

In this step:
* A local SQLite database is created  
* Two tables are defined:
  * `customers` - customer profile and account information  
  * `transactions` - individual customer purchase records  
* Data is loaded from CSV files into the database  

Using SQL at this stage makes it easier to validate the data structure and perform reliable aggregations later in the analysis.

```python
def create_database():
    if not os.path.exists('database'):
        os.makedirs('database')

    conn = sqlite3.connect('database/churn.db')
    cursor = conn.cursor()

    cursor.execute('DROP TABLE IF EXISTS transactions')
    cursor.execute('DROP TABLE IF EXISTS customers')

    cursor.execute('''
        CREATE TABLE customers (
            customer_id INTEGER PRIMARY KEY,
            signup_date TEXT,
            region TEXT,
            plan_type TEXT,
            is_active INTEGER
        )
    ''')
    print("Created customers table")

    cursor.execute('''
        CREATE TABLE transactions (
            transaction_id INTEGER PRIMARY KEY,
            customer_id INTEGER,
            transaction_date TEXT,
            transaction_amount REAL,
            FOREIGN KEY (customer_id) REFERENCES customers (customer_id)
        )
    ''')
    print("Created transactions table")

    try:
        with open('data/customers.csv', 'r') as f:
            next(f)
            for line in f:
                cursor.execute('INSERT INTO customers VALUES (?, ?, ?, ?, ?)', line.strip().split(','))
        print("Inserted customer data")

        with open('data/transactions.csv', 'r') as f:
            next(f)
            for line in f:
                cursor.execute('INSERT INTO transactions VALUES (?, ?, ?, ?)', line.strip().split(','))
        print("Inserted transaction data")

    except FileNotFoundError:
        print("WARNING: CSV files not found in data/ directory")

    conn.commit()
    conn.close()
    print("Database setup complete")

create_database()
```

## 3. Data Loading

### What’s happening here

Customer and transaction data are loaded from the SQLite database into Pandas DataFrames.

This reflects a common analytics workflow:
* **SQL** is used for storing and querying data  
* **Python** is used for analysis, feature creation, and visualization  

Once the data is in Pandas, it becomes easier to explore patterns and prepare it for further analysis.

```python
def get_connection():
    return sqlite3.connect('database/churn.db')

customers = pd.read_sql('SELECT * FROM customers', get_connection())
transactions = pd.read_sql('SELECT * FROM transactions', get_connection())

print(f"Loaded {len(customers)} customers")
print(f"Loaded {len(transactions)} transactions")

customers.head()
```

## 4. SQL Analysis

### What this section focuses on

Before doing any modeling or feature creation, SQL is used to compute a few high-level churn statistics.

### Questions being answered
* How many customers are currently active?  
* How many customers have churned?  

This step provides a quick snapshot of the overall churn situation and helps guide the rest of the analysis.

```python
# Calculate Overall Churn
query_churn = '''
SELECT is_active, COUNT(*) AS count
FROM customers
GROUP BY is_active
'''
print("Overall Churn Stats:")
print(pd.read_sql(query_churn, get_connection()))

# Calculate Churn by Region (Needed for visualization)
query_region = '''
    SELECT 
        region,
        COUNT(*) as total_customers,
        SUM(CASE WHEN is_active = 0 THEN 1 ELSE 0 END) as churned_customers,
        ROUND(100.0 * SUM(CASE WHEN is_active = 0 THEN 1 ELSE 0 END) / COUNT(*), 2) as churn_rate_percent
    FROM customers
    GROUP BY region
    ORDER BY churn_rate_percent DESC
'''
churn_by_region = pd.read_sql(query_region, get_connection())
print("\nChurn by Region calculated.")
churn_by_region
```

## 5. Customer Metrics (Feature Creation)

### What’s done here

Transaction-level data is aggregated into customer-level metrics, including:
* Total spending per customer  
* Average transaction amount  
* Number of transactions  
* Customer tenure (measured in days)  

### Why this step is important

Most business decisions are made at the **customer level**, not at the transaction level.  
This step converts raw transaction data into meaningful customer metrics that can be used for churn analysis and risk scoring.

```python
features = transactions.groupby('customer_id').agg(
    total_spending=('transaction_amount', 'sum'),
    avg_transaction=('transaction_amount', 'mean'),
    transaction_count=('transaction_amount', 'count')
).reset_index()

customers['signup_date'] = pd.to_datetime(customers['signup_date'])
features = customers.merge(features, on='customer_id', how='left').fillna(0)

reference_date = datetime(2025, 8, 1)
features['term_days'] = (reference_date - features['signup_date']).dt.days

features.head()
```

## 6. Churn Risk Modeling
### Purpose of this section

A simple rule-based churn risk score is calculated instead of using complex machine learning models.

### Risk score logic:
* Fewer transactions -> higher churn risk  
* Lower spending -> higher churn risk  
* Shorter customer tenure -> slightly higher churn risk  

### Why rule-based?

* Easy to interpret  
* Transparent logic  
* Appropriate for early-stage analysis and intern-level projects  

```python
median_spending = features['total_spending'].median()

features['churn_risk'] = (
    np.where(features['transaction_count'] < 3, 0.4, 0.1) +
    np.where(features['total_spending'] < median_spending, 0.3, 0.1) +
    np.where(features['term_days'] < 100, 0.2, 0.05)
).clip(0, 1)

features[['customer_id', 'is_active', 'churn_risk']].head()
```

## 6. Churn Risk Modeling

### Purpose of this section

Instead of using a complex machine learning model, this project applies a **simple rule-based churn risk score**.

### How the risk score works
* Customers with fewer transactions are assigned higher risk  
* Customers with lower spending are assigned higher risk  
* Customers with shorter account tenure are assigned slightly higher risk  

### Why a rule-based approach was chosen
* The logic is easy to understand and explain  
* Results are transparent and interpretable  
* This approach is well-suited for early-stage analysis and an intern-level project  

The goal here is clarity and reasoning, not model complexity.

```python
# Plot 1: Total Spending vs Churn
plt.figure(figsize=(8, 6))

spending_summary = features.groupby('is_active')['total_spending'].mean()
bars = plt.bar(['Churned', 'Active'], [spending_summary[0], spending_summary[1]], color=['#285700', '#1871BA'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, f'${height:.2f}', ha='center', va='bottom', fontsize=11)

plt.title('Average Total Spending by Churn Status', fontsize=14, fontweight='bold')
plt.ylabel('Average Total Spending ($)', fontsize=11)
plt.xlabel('Customer Status', fontsize=11)
plt.grid(axis='y', alpha=0.333)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Number of Transactions vs Churn
plt.figure(figsize=(8, 6))

transaction_summary = features.groupby('is_active')['transaction_count'].mean()
bars = plt.bar(['Churned', 'Active'], [transaction_summary[0], transaction_summary[1]], color=['#285700', '#1871BA'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.1f}',ha='center', va='bottom', fontsize=11)

plt.title('Average Transaction Count by Churn Status', fontsize=14, fontweight='bold')
plt.ylabel('Average Number of Transactions', fontsize=11)
plt.xlabel('Customer Status', fontsize=11)
plt.grid(axis='y', alpha=0.333)
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Churn Rate by Region
plt.figure(figsize=(10, 6))
# FIXED: Changed variable name from 'churn_by_region_df' to 'churn_by_region'
bars = plt.bar(churn_by_region['region'], churn_by_region['churn_rate_percent'], color='#285700')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, f'{height:.1f}%', ha='center', va='bottom', fontsize=11)

plt.title('Churn Rate by Region', fontsize=14, fontweight='bold')
plt.ylabel('Churn Rate (%)', fontsize=11)
plt.xlabel('Region', fontsize=11)
plt.grid(axis='y', alpha=0.333)
plt.ylim(0, max(churn_by_region['churn_rate_percent']) * 1.2)
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Churn Risk Score Distribution
plt.figure(figsize=(10, 6))

active_risks = features[features['is_active'] == 1]['churn_risk']
churned_risks = features[features['is_active'] == 0]['churn_risk']

plt.hist([active_risks, churned_risks], bins=10, label=['Active', 'Churned'], color=['#1871BA', '#285700'])

plt.title('Distribution of Churn Risk Scores', fontsize=14, fontweight='bold')
plt.xlabel('Churn Risk Score', fontsize=11)
plt.ylabel('Number of Customers', fontsize=11)
plt.legend(fontsize=11)
plt.grid(axis='y', alpha=0.333)
plt.tight_layout()
plt.show()
```
